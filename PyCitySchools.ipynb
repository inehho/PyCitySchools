```python
import pandas as pd
import numpy as np
```

```python
#read in file
file1 = "Resources/students_complete.csv"
students_df = pd.read_csv(file1)
students_df.head()
```

```python

#read in file2
file2 = "Resources/schools_complete.csv"
schools_df = pd.read_csv(file2)
schools_df.head()
```

```python

#combining data files
school_district = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
school_district.head()
```

```python
num_school = school_district["school_name"].nunique() #nunique returns the no of unique values
num_school

num_stu = school_district["Student ID"].nunique()
num_stu

#total budget calculation
total_budget= school_district["budget"].unique().sum()
total_budget

#avg math score for the district
avg_mathscore = school_district["math_score"].mean()
avg_mathscore

#avg reading score for the district
avg_reading= school_district["reading_score"].mean()
avg_reading

#avg passrate for the district
passing_rate = (avg_mathscore + avg_reading) /2
passing_rate

#%maths pass rate
Math_Pass_ratedf = school_district.loc[school_district["math_score"] >= 70]
Math_Pass_ratedf= (Math_Pass_ratedf ["math_score"].count()/num_stu) *100

Math_Pass_ratedf

#% reading pass rate
Reading_pass_ratedf = school_district.loc[school_district["reading_score"] >= 70]
Reading_pass_ratedf= (Reading_pass_ratedf ["reading_score"].count()/num_stu) *100

Reading_pass_ratedf
```

```python
New_district_results= {"Total Number of Schools": (num_school),
                       "Total Number of Students": (num_stu),
                       "Total Budget": (total_budget),
                       "Average Math Score": (avg_mathscore),
                       "Average Reading Score": (avg_reading),
                       "% Passing Math": (Math_Pass_ratedf),
                       "% Passing Reading": (Reading_pass_ratedf),
                       "Overall Passing Rate": (passing_rate)}
District_results = pd.DataFrame(New_district_results, index=[0])
District_results.head()
```

```python
#Creating an overview of a table which summarizes key metrics of each school
#assembling all data

#count of students per school
students_count= school_district.groupby(["school_name"]).count()["Student ID"]
students_count.head()

#total budget per school
total_school_budget = school_district.groupby(["school_name"]).mean()["budget"]
total_school_budget.head()

#size per school
school_size = school_district.groupby(["school_name"]).mean()["size"]
school_size.head()

#student budget per school
per_student_budget = total_school_budget / school_size
per_student_budget.head()

#avg math score per school
avg_math_score= school_district.groupby(["school_name"]).mean()["math_score"]
avg_math_score

#avg reading score per school
avg_reading_score= school_district.groupby(["school_name"]).mean()["reading_score"]
avg_reading_score

#percentage of math pass rate per school
school_pass_rate_math = school_district.loc[school_district["math_score"] >= 70]
school_pass_rate_math = (school_pass_rate_math.groupby(["school_name"]).count()["math_score"])/(school_district.groupby(["school_name"]).count()["size"])
school_pass_rate_math= school_pass_rate_math * 100

school_pass_rate_math.head()

#percentage of reading pass rate per school
school_reading_overall = school_district.loc[school_district["reading_score"] >= 70]
school_reading_overall = (school_reading_overall.groupby(["school_name"]).count()["reading_score"])/(school_district.groupby(["school_name"]).count()["size"])
school_reading_overall = school_reading_overall * 100
school_reading_overall.head()


overall_pass_rate = (school_pass_rate_math + school_reading_overall)/2
overall_pass_rate.head()
```

```python
topschools= {"Total Students": students_count,
              "Total School Budget": total_school_budget,
              "Per Student Budget": per_student_budget,
              "Average Math Score": avg_math_score,
              "Average Reading Score": avg_reading_score,
              "% Passing Math": school_pass_rate_math,
              "% Passing Reading": school_reading_overall,
              "% Overall Passing Rate": overall_pass_rate
             }

topschools_df = pd.DataFrame(topschools)
topschools_df.head()
```

```python
topschools_df= topschools_df.sort_values(["% Overall Passing Rate"], ascending = False)
topschools_df.head()
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_district['spending_bins'] = pd.cut(school_district['budget']/school_district['size'], spending_bins, labels= group_names)

by_spending =school_district.groupby('spending_bins')

#calculations
#avg math score per school
avg_math= by_spending["math_score"].mean()
avg_math

#avg reading score per school
avg_reading= by_spending["reading_score"].mean()
avg_reading

#percentage of math pass rate per school
pass_math = school_district[school_district['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_math= pass_math*100
pass_math.head()

#percentage of reading pass rate per school
reading_overall = school_district[school_district["reading_score"] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
reading_overall= reading_overall* 100
reading_overall.head()


overall = school_district[(school_district['reading_score'] >= 70) & (school_district['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall= overall * 100

scores_by_spending= {"Average Math Score": avg_math,
              "Average Reading Score": avg_reading,
              "% Passing Math": pass_math,
              "% Passing Reading": reading_overall,
              "% Overall Passing Rate": overall
             }


scores_by_spending = pd.DataFrame(scores_by_spending)
scores_by_spending.head()
scores_by_spending['% Overall Passing Rate'] = (scores_by_spending['% Passing Math'] + scores_by_spending['% Passing Reading']) / 2
scores_by_spending.index.name = "Spending Ranges (Per Student Budget)"
scores_by_spending = scores_by_spending.reindex(group_names)
scores_by_spending.head()
```

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_district['size_bins'] = pd.cut(school_district['size'], size_bins, labels= group_names)

by_size =school_district.groupby('size_bins')

#calculations
#avg math score by school size
avg_math= by_size["math_score"].mean()
avg_math

#avg reading score per school size
avg_reading= by_size["reading_score"].mean()
avg_reading

#percentage of math pass rate by school size 
pass_math = school_district[school_district['math_score'] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
pass_math= pass_math*100
pass_math.head()

#percentage of reading pass rate by school size
reading_overall = school_district[school_district["reading_score"] >= 70].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
reading_overall= reading_overall* 100
reading_overall.head()


overall = school_district[(school_district['reading_score'] >= 70) & (school_district['math_score'] >= 70)].groupby('size_bins')['Student ID'].count()/by_size['Student ID'].count()
overall= overall * 100

scores_by_size= {"Average Math Score": avg_math,
              "Average Reading Score": avg_reading,
              "% Passing Math": pass_math,
              "% Passing Reading": reading_overall,
              "% Overall Passing Rate": overall
             }


scores_by_size = pd.DataFrame(scores_by_spending)
scores_by_size.head()
scores_by_size['% Overall Passing Rate'] = (scores_by_size['% Passing Math'] + scores_by_size['% Passing Reading']) / 2
scores_by_size.index.name = "School Size"
scores_by_size = scores_by_size.reindex(group_names)
scores_by_size.head()
```
